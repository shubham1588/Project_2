## Find Outliers

```python
Q1=Dataset_3[['temp','hum','atemp']].quantile(.25)
Q3=Dataset_3[['temp','hum','atemp']].quantile(.75)
```

```python
def find_outliers(Dataset_3,cols):
    Q1=Dataset_3[cols].quantile(.25)
    Q3=Dataset_3[cols].quantile(.75)
    iqr=Q3-Q1
    min_r=Q1-1.5*iqr
    max_r=Q3+1.5*iqr
    outlier_indices=Dataset_3.index[(Dataset_3[cols]<min_r)|(Dataset_3[cols]>max_r)]
    return outlier_indices
```

```python
find_outliers(Dataset_3,'atemp')
```

```python
find_outliers(Dataset_3,'temp')
```

```python
find_outliers(Dataset_3,'hum')
```
